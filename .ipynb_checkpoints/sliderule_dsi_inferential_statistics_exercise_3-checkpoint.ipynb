# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

### Introduction
> Define functions

```python
def draw_bs_pairs(x, y, func, size=1):
    """Perform pairs bootstrap for linear regression."""

    # Set up array of indices to sample from: inds
    inds = np.arange(len(x))
    
    # Set up x array for regression
    x_array = np.array([0,max(x)])

    # Initialize replicates: bs_slope_reps, bs_intercept_reps
    bs_slope_reps = np.empty(size)
    bs_intercept_reps = np.empty(size)
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_inds = np.random.choice(inds, size=len(inds))
        bs_x, bs_y = x[bs_inds], y[bs_inds]
        bs_slope_reps[i], bs_intercept_reps[i] = np.polyfit(bs_x, bs_y, 1)
        bs_replicates[i] = func(bs_x, bs_y)
    
    return bs_replicates, bs_slope_reps, bs_intercept_reps
```

```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat = np.corrcoef(x,y)

    # Return entry [0,1]
    return corr_mat[0,1]
```

```python
def mean_excess(data, size=1):
    '''
    Compute the mean excess readmission ratio for number of discharges
        (<100) and (>1000)
    '''
    # Set up array of indices to sample from: inds
    inds = np.array(data.index)
    # Initialize replicates: under_100, over_1000
    under_100_mean = np.empty(size)
    under_100_percent = np.empty(size)
    over_1000_mean = np.empty(size)
    over_1000_percent = np.empty(size)
    mean_diff = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_inds = np.random.choice(inds, size=len(inds))
        bs_data = data.reindex(bs_inds)

        # Data under 100 discharges
        data_under_100 = bs_data[bs_data['Number of Discharges'] < 100]
        data_under_100 = data_under_100.dropna(subset=['Excess Readmission Ratio'])
        under_100_mean[i] = data_under_100['Excess Readmission Ratio'].mean()
        under_100_percent[i] = np.sum(data_under_100['Excess Readmission Ratio'] > 1.0) / len(data_under_100)
        
        # Data over 1000 discharges
        data_over_1000 = bs_data[bs_data['Number of Discharges'] > 1000]
        data_over_1000 = data_over_1000.dropna(subset=['Excess Readmission Ratio'])
        over_1000_mean[i] = data_over_1000['Excess Readmission Ratio'].mean()
        over_1000_percent[i] = np.sum(data_over_1000['Excess Readmission Ratio'] > 1.0) / len(data_over_1000)
        
        # Mean difference
        mean_diff[i] = under_100_mean[i] - over_1000_mean[i]
    
    return under_100_mean, under_100_percent, over_1000_mean, over_1000_percent, mean_diff
```

### A. Do you agree with the above analysis and recommendations? Why or why not?

Furthermore, it appears as though there aren't enough data points for hospitals with large capacity (number of discharges >1000). It seems that if there were more data points for large capacity hospitals, there would not be much of a difference between mean excess readmission rate for hospitals with small capacity (number of discharges <100) and hospitals with large capacity (number of discharges >1000).

```python
# Create DataFrame
clean = clean_hospital_read_df[['Number of Discharges', 'Excess Readmission Ratio']]
x = np.array(clean_hospital_read_df['Number of Discharges'][81:-3])
y = np.array(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])
slope, inter = np.polyfit(x,y,1)
x1 = np.array([0,max(x)])
y1 = slope * x1 + inter
```

```python
obs_corr = pearson_r(x, y)
print('Observed Pearson Correlation:', obs_corr)

under_100 = clean[clean['Number of Discharges'] < 100]
under_100 = under_100.dropna(subset=['Excess Readmission Ratio'])
under_100_obs_mean = under_100['Excess Readmission Ratio'].mean()
under_100_obs_percent = np.sum(under_100['Excess Readmission Ratio'] > 1.0) / len(under_100)
print('Under 100, Observed Mean Excess Ratio:', under_100_obs_mean)
print('Under 100, Observed Excess Percentage:', under_100_obs_percent)

over_1000 = clean[clean['Number of Discharges'] > 1000]
over_1000 = over_1000.dropna(subset=['Excess Readmission Ratio'])
over_1000_obs_mean = over_1000['Excess Readmission Ratio'].mean()
over_1000_obs_percent = np.sum(over_1000['Excess Readmission Ratio'] > 1.0) / len(over_1000)
print('Over 1000, Observed Mean Excess Ratio:', over_1000_obs_mean)
print('Over 1000, Observed Excess Percentage:', over_1000_obs_percent)

obs_mean_diff = under_100_obs_mean - over_1000_obs_mean
print('Observed Difference:', obs_mean_diff)

plt.plot(x,y,marker='.',linestyle='none')
plt.plot(x1,y1)
plt.xlabel('Number of discharges')
plt.ylabel('Excess rate of readmissions');
```

### B. Provide support for your arguments and your own recommendations with a statistically sound analysis:
1. Setup an appropriate hypothesis test.

  - Construct another plot that conveys the same information in a more direct manner.
> I have plotted a histogram for a better comparison.

```python
bs_replicates, bs_slope_reps, bs_intercept_reps = draw_bs_pairs(x, y, pearson_r, size=1000)
# Compute p-value
p_corr = np.sum(bs_replicates >= obs_corr) / len(bs_replicates)
print('Correlation p-value:', p_corr)

# Plot linear regressions
plt.plot(x, y, marker='.', linestyle='none')
for i in range(1000):
    plt.plot(x, bs_slope_reps[i]* x + bs_intercept_reps[i],
             linewidth=0.5, alpha=0.2, color='red')
plt.xlabel('Number of discharges')
plt.ylabel('Excess rate of readmissions');

under_100_mean, under_100_percent, over_1000_mean, over_1000_percent, mean_diff = mean_excess(clean, size=10000)
# Compute p-values
p_under_100_mean = np.sum(under_100_mean >= under_100_obs_mean) / len(under_100_mean)
p_under_100_percent = np.sum(under_100_percent >= under_100_obs_percent) / len(under_100_percent)
p_over_1000_mean = np.sum(over_1000_mean >= over_1000_obs_mean) / len(over_1000_mean)
p_over_1000_percent = np.sum(over_1000_percent >= over_1000_obs_percent) / len(over_1000_percent)
p_mean_diff = np.sum(mean_diff >= obs_mean_diff) / len(mean_diff)
print('Under 100, Mean Excess Ratio p-value:', p_under_100_mean)
print('Under 100, Excess Percentage p-value:', p_under_100_percent)
print('Over 1000, Mean Excess Ratio p-value:', p_over_1000_mean)
print('Over 1000, Excess Percentage p-value:', p_over_1000_percent)
print('Mean Difference p-value:', p_mean_diff)
```
